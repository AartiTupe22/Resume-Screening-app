```python
#import libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import pickle
import spacy
```

```python
#load spacy english model
nlp = spacy.load('en_core_web_sm')
```

```python
#load the dataset
abc = pd.read_csv('UpdatedResumeDataSet.csv')
```

```python
abc.head(5)
```

```python
#define a function to preprocess text
def preprocess_text(text):
    # Create a spaCy document object
    doc = nlp(text)

    # Clean the text by removing stop words, punctuation, and non-alphabetic characters
    clean = []
    for token in doc:
        if not token.is_stop and not token.is_punct and token.is_alpha:
            clean.append(token.lemma_.lower())
    # Return the cleaned text as a string
    return " ".join(clean)
     
```

```python
#apply preprocess text as reume column and store the result in new column
abc['resume_text'] = abc['Resume'].apply(preprocess_text)
```

```python
label_encoder = LabelEncoder()
```

```python
#encode category column
abc['Category_encoded'] = label_encoder.fit_transform(abc['Category'])
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(abc['resume_text'], abc['Category_encoded'], test_size=0.20, random_state=2, stratify=abc['Category'])
```

```python
# Create a pipeline consisting of a TfidfVectorizer and a KNeighborsClassifier
model =Pipeline([('vectorizer', TfidfVectorizer()), ('model', KNeighborsClassifier())])
```

```python

# Fit the model on the training data
model.fit(X_train, y_train)
```

```python
# Predict the labels of the test data
y_pred = model.predict(X_test)
```

```python
# Print the classification report
print(classification_report(y_test, y_pred))
```

```python
# Compute and display the confusion matrix
cm_matrix = confusion_matrix(y_test, y_pred)

ax = plt.subplot()
sns.heatmap(cm_matrix, annot=True, ax=ax)

ax.set_xlabel('Predicted labels', fontsize=12)
ax.set_ylabel('True labels', fontsize=12)
ax.set_title('Confusion Matrix', fontsize=14)

plt.show()
```

```python

# Print the accuracy score
print(accuracy_score(y_test,y_pred))
```

```python

# # Save the model to a file
with open('model_resume.pkl', 'wb') as f:
    pickle.dump(model, f)
     
```
